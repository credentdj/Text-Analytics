```python
import sys
import json
import re
import string
import nltk
import uuid
import glob
import numpy
from numpy import mean, ptp, var, std, median
from scipy.stats import mode
```

```python
files = glob.glob('C:\Users\yxiao\Desktop\High_Output2\*.json')
```

```python
for passage in files:
    content = json.load(open(passage))
    sen_length = []
    for sentence in content:
        length = len(sentence["sentence"])
        sen_length.append(length)
    variance = var(sen_length)
    standard = std(sen_length)
    Average_length = mean(sen_length)
    Median = median(sen_length)
    Mode = mode(sen_length)
    print Mode
```
