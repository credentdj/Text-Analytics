```python
import nltk
from nltk import FreqDist
import sys
import json
import re
import string
import nltk
import uuid
import glob
import numpy
from numpy import mean, ptp, var, std, median
from scipy.stats import mode
from nltk.corpus import stopwords
```

```python
files = glob.glob('C:\Users\yxiao\Desktop\High_Output2\*.json')
```

```python
for passage in files: 
    raw = json.load(open(passage))
    sen_word_len = []
    for sentence in raw:
        sentence_raw = sentence["sentence"]
        tokens = nltk.word_tokenize(sentence_raw)
        word_num = len(tokens)-1
        sen_word_len.append(word_num)

    Mean = mean(sen_word_len)
    Variance = var(sen_word_len)
    Standard = std(sen_word_len)
    Median = median(sen_word_len)
    Mode = mode(sen_word_len)

    print Median
#     print Variance
#     print Standard
#     print Median
```
